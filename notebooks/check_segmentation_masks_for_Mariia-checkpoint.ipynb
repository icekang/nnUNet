```python
from pathlib import Path
import torchio as tio
```

```python
segmentation_mask_all_dir = Path('/home/gridsan/nchutisilp/datasets/Mariia/P3_MIT/post_pci_prediction_LaW_all')
segmentation_mask_merged_dir = Path('/home/gridsan/nchutisilp/datasets/Mariia/P3_MIT/post_pci_prediction_LaW_merged304_all')
```

```python
trained_on_all_subjects = dict()
for nii_path in segmentation_mask_all_dir.glob("*.nii.gz"):
    trained_on_all_subjects[nii_path.name] = tio.LabelMap(nii_path)


trained_on_merged_all_subjects = dict()
for nii_path in segmentation_mask_merged_dir.glob("*.nii.gz"):
    trained_on_merged_all_subjects[nii_path.name] = tio.LabelMap(nii_path)
```

```python
samples_of_trained_on_all_subjects = {k: v for k, v in list(trained_on_all_subjects.items())[:10]}
tio.Subject(samples_of_trained_on_all_subjects).plot(figsize=(40, 20))
```

```python
samples_of_trained_on_merged_all_subjects = {k: v for k, v in list(trained_on_merged_all_subjects.items())[:10]}
tio.Subject(samples_of_trained_on_merged_all_subjects).plot(figsize=(40, 20))
```
